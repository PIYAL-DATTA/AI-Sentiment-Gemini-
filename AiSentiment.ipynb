```python
import time
from google.api_core.exceptions import TooManyRequests
import csv

# Save to CSV
df.to_csv("filtered_data.csv", index=False)

data = []
with open('filtered_data.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        data.append(row)

count = 0
temp = "{\'im getting on borderlands and i will murder you all ,': \'"

# Now, you can iterate through your data and create prompts
# im getting on borderlands and i will murder you all
for x in data:
    p = str(x)
    test = p[len(temp):len(p)-3]
    prompt = f"{test}.\n Tell me if it is positive, negative or neutral. Don't need any explanation or anything else. Just reply positive, negative or neutral."

    # Error handling and retry logic
    retries = 3  # Number of retry attempts
    delay = 20  # Delay in seconds between retries

    for i in range(retries):
        try:
            response = model.generate_content(prompt)
            count = count + 1
            print(f"Prompt: {prompt}\nResponse: {response.text}\n---  {count}")
            break  # Exit the retry loop if successful
        except TooManyRequests:
            if i < retries - 1:
                print(f"Rate limit exceeded. Retrying in {delay} seconds...")
                time.sleep(delay)
            else:
                print(f"Rate limit exceeded. Giving up after {retries} retries.")
                raise  # Re-raise the exception if all retries fail

    # Introduce a delay to avoid hitting the rate limit
    time.sleep(2)  # Adjust the delay as needed
```
